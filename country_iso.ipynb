```python
import pandas as pd
import numpy as np
import pycountry
```

```python
df_qs = pd.read_excel('qs_ranking.xlsx')
df_qs
```

```python
# 先将Taiwan, Hong Kong SAR, Macau SAR, China (Mainland)替换为China
df_qs['所处国家'] = df_qs['所处国家'].replace('Taiwan', 'China')
df_qs['所处国家'] = df_qs['所处国家'].replace('Hong Kong SAR', 'China')
df_qs['所处国家'] = df_qs['所处国家'].replace('Macau SAR', 'China')
df_qs['所处国家'] = df_qs['所处国家'].replace('China (Mainland)', 'China')
```

```python
# 将所处国家的全称转换为ISO代码
def country_to_iso(country):
    try:
        return pycountry.countries.lookup(country).alpha_3
    except LookupError:
        return np.nan
    
df_qs['iso'] = df_qs['所处国家'].apply(country_to_iso)

df_qs
```

```python
# 查找是否有缺失值
df_qs[df_qs['iso'].isnull()]['所处国家'].unique()
```

```python
# 把'Russia', 'Turkey', 'Brunei'分别填充为'RUS', 'TUR', 'BRN'
df_qs.loc[df_qs['所处国家'] == 'Russia', 'iso'] = 'RUS'
df_qs.loc[df_qs['所处国家'] == 'Turkey', 'iso'] = 'TUR'
df_qs.loc[df_qs['所处国家'] == 'Brunei', 'iso'] = 'BRN'
```

```python
# 把‘所处国家”列的列名改为‘Country’
df_qs = df_qs.rename(columns={'所处国家': 'Country'})
```

```python
df_iso_region = pd.read_csv('iso_countries_region.csv')
df_iso_region.head(5)
```

```python
# 将df_qs中新建一列，为region列，根据df_iso_region中的数据填充
df_qs['Region'] = df_qs['iso'].map(df_iso_region.set_index('alpha-3')['region'])
df_qs['Sub-region'] = df_qs['iso'].map(df_iso_region.set_index('alpha-3')['sub-region'])
```

```python
# 导出数据
df_qs.to_excel('qs_ranking_iso.xlsx', index=False)
```
